## Exponential Growth - Like doubling grains of rice on a chess board
Why are people worried about COVID-19?  Consider this thought experiment.  How much rice would you need if you doubled the number of grains of rice on successive squares of a chessboard?  Put one grain of rice on the first square, two on the next, four on the one after that, and so on. In the end, there would be 9,223,372,036,854,775,808 grains of rice in the last square. There would be 18,446,744,073,709,551,615 grains of rice on the chessboard in total. Now that's a lot of rice! Don't believe it?  Check the math for yourself below.

So what does this mean for COVID-19?  If one infected person infects two others, and each of them infect two others, and so on and so on and so on, then a lot of people end up getting infected.  This is why the authorities are worried. Rather than being frightened, though, take action and responsibility to safeguard your health, and that of your family, neighbors, and vulnerable people in the community. This site will help you to do so!
